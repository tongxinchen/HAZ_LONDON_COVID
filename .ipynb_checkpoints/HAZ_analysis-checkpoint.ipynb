# 1 HAZ Clustering
__cluster labels and silhouette score results are in /data/haz_result__

```python
def HCluster(df,k=5,linkstr='ward',affinity='euclidean'):
    """hclustering and get cluster labels"""
    df_union_cluster=df.copy()
    df_union_cluster=df_union_cluster.replace({np.nan: -1})
    
    HCluster = AgglomerativeClustering(n_clusters=k, linkage=linkstr).fit(df_union_cluster)
    
    return HCluster

def evaluate_num_of_clusters(df,nclusters=range(2,10),linkage='ward',aff='euclidean'):
    """optimization of cluster numbers"""
    scores=[]
    for k in nclusters:
        score = silhouette_score(df, HCluster(df,k,linkstr=linkage,affinity=aff).labels_, metric='euclidean')
        scores.append(score)
    return nclusters,scores
```

```python
range_name = ['Before lockdown','First national lockdown','Minimal lockdown restrictions',
              'Reimposing restrictions','Three-tier restriction','Second national lockdown',
              'Four-tier restriction','Third national lockdown','All observation time periods']
date_start = ['2020-01-01','2020-03-23','2020-07-04','2020-09-14','2020-10-14','2020-11-05','2020-12-03','2021-01-06','2020-01-01']
date_end = ['2020-03-22','2020-07-03','2020-09-13','2020-10-13','2020-11-04','2020-12-02','2021-01-05','2021-02-27','2021-02-27']

range_name_test = ['Before lockdown (test)']
date_start_test = ['2020-01-01']
date_end_test = ['2020-01-31']

df_ff_2020 =  df_ff_2020_sample.iloc[0:100,:] # we select 100 oas for test as computation at 23053 oas 

for i in range(1): # 
    # 1 row based standardized
    print(range_name_test[i])
    df_ff_2020_s = df_ff_2020.T.loc[date_start_test[i]:date_end_test[i]].T
    print(df_ff_2020_s.shape)
    patterns_data = df_ff_2020_s
    # # do not execute as sample data are row-based standardised footfalls 
    # x = (patterns_data.transpose() - patterns_data.transpose().mean())/ (patterns_data.transpose().std()) 
    # patterns_data = x.transpose()
    
    # 2 optimization of cluster numbers
    affstr = ["euclidean", "manhattan", "cosine"]
    fig, ax = plt.subplots(1, len(affstr), figsize=(15, 5))
    count = 0
    for affstr_ in affstr:
        nclusters, sil_scores = evaluate_num_of_clusters(patterns_data, nclusters=range(5, 20), linkage='ward',aff=affstr_)
        sns.scatterplot(x=nclusters, y=sil_scores, ax=ax[count])
        sns.lineplot(x=nclusters, y=sil_scores, ax=ax[count], style=True, dashes=[(2, 2)])
        ax[count].set_xlabel('Cluster numbers', fontsize=12)
        ax[0].set_ylabel('Silouette score', fontsize=12)
        ax[count].set_title(affstr_)
        count += 1
    
    # 3 plot cluster numbers vs socres 
    fig.suptitle('Linkage:Ward', fontsize=15)
    df_socre = pd.DataFrame({'sil_scores': sil_scores, 'n_cluster': [i for i in range(5, 20)]})
    df_socre.to_csv(f'data/test/y_sil_scores_{i}.csv')
    plt.show()

    # 4 detemine the cluster numbers and get cluster labels
    n_cluster = df_socre[df_socre.sil_scores == df_socre.sil_scores.max()].n_cluster.values[0]
    print('cluster numbers', n_cluster)
    res = HCluster(patterns_data, k= n_cluster , linkstr='ward', affinity='euclidean')
    
    # 5 plot cluster hist
    plt.hist(res.labels_, color='royalblue')
    plt.show()
    pd.DataFrame({'labels': res.labels_ }).to_csv(f'data/test/y_labels_{i}.csv')
```

```python
def get_dict_labels(df_ff_mean):
    """get new cluster labels dict ranked by the daily average footfall"""
    
    df_ff_mean = df_ff_mean.sort_values(by = 'v_mean', ascending = False)
    df_ff_mean['new_labels'] = [i for i in range(len(df_ff_mean))]
    dict_labels = dict(zip(df_ff_mean.lables.to_list(), df_ff_mean.new_labels.to_list() ))
    
    return dict_labels
```

```python
def get_map_fig(df_new_labels, gdf_london, colors, i):  
    """plot HAZ map"""
    
    import matplotlib.patches as mps
    fig,ax = plt.subplots(figsize=(6,6))
    path_list =[]
    
    for c_type, data in gdf_london.groupby('new_labels'): 
        
        print('haz plot', c_type)
        color = colors[c_type]    
        data.plot(color=color,ax=ax,label = c_type, alpha = 1, lw =0.00000001)
        ax.axis('off')
    
        p_i = mps.Patch(color = colors[c_type], label =f'HAZ {c_type} : ' + str(df_new_labels.query('new_labels=='+str(c_type)).shape[0])+' OAs')
        path_list.append(p_i)
 
    
    ax.legend(handles = path_list,fontsize = 13, frameon =False, loc = [0, -0.3], ncol = 2 )
    ax.set_title(f'{range_name[i]} \n {date_start[i]} -- {date_end[i]}', fontsize = 21, fontweight='bold')
    
    plt.savefig(f'data/figs/haz_map_{i}.jpg',dpi = 200, bbox_inches='tight')
```

```python
def get_haz_time_fig(df_new_labels, colors, i):
    """plot HAZ temporal pattern"""
    
    from matplotlib.ticker import FormatStrFormatter    
    fig,ax=plt.subplots(len(df_new_labels.new_labels.unique()), 1, figsize = (6, 0.6*len(df_new_labels.new_labels.unique())))     
    
    for k in np.sort(df_new_labels.new_labels.unique()):
        print('HAZ footfall pattern', k)
     
        df_plot = pd.read_csv(f'data/plot/op_{i}_label_{k}.csv')
        days_prev = len(df_plot)
        data_plot = df_plot.iloc[:,1].to_list()
        
        ax[k].plot(range(0,days_prev), data_plot , marker='', color='black', linewidth = 1.5, alpha = 0.8)
        ax[k].plot(range(0,days_prev), data_plot , marker='', color=colors[k], linewidth = 1, alpha = 0.95)        
        we = [i for i, n in enumerate(list(pd.to_datetime(df_plot.iloc[:,0].to_list()).dayofweek)) if n == 5 or n == 6]
        
        for j in we:
            ax[k].axvline(j, color ='black',linestyle = '-', lw = 0.4) 
        ax[k].set_ylim([ np.min(data_plot) - 0.1 , np.max(data_plot) + 0.1])        
        ax[k].set_yticks([ np.min(data_plot) - 0.1 , np.max(data_plot) + 0.1])
        ax[k].set_yticklabels([ np.min(data_plot) - 0.1 , np.max(data_plot) + 0.1] ,rotation = 0, horizontalalignment='right',fontsize=13)        
        ax[k].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))       
        ax[k].tick_params(axis='y', labelsize=13)
        ax[k].tick_params(axis='x',  
                          which='both',  
                          bottom=False,  
                          top=False,  
                          labelbottom=False)       
    
    ax[-1].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))    
    ax[-1].tick_params(axis='y', labelsize=13)    
    ax[-1].tick_params(
            axis='x',  
            which='both', 
            bottom= True, 
            top=False,  
            labelbottom=True)    
    
    ax[-1].set_ylim([ np.min(data_plot) - 0.1 , np.max(data_plot) + 0.1])     
    ax[-1].set_xticks([0, days_prev-1])    
    ax[-1].set_xticklabels([date_start[i], date_end[i]] ,rotation = 0, horizontalalignment='center',fontsize=18)
        
    plt.tight_layout(h_pad=0.0001)
    sns.despine(bottom=True, left=False)
    ax[-1].spines.bottom.set_visible(True)

    plt.savefig(f'data/figs/ts_{i}.jpg',dpi = 300, bbox_inches='tight')
```

```python
# evaluate rf model
from sklearn import preprocessing
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
# define dataset
df_london_lad_use_new = pd.read_csv('data/df_london_lad_use_new.csv', index_col = 0)

all_accuracy_list = []
for n in df_london_lad_use_new.columns.to_list()[11:20]:
    print(n)
    X = pd.DataFrame(preprocessing.scale(df_london_lad_use_new.iloc[:,1:11]))
    X['supergroup'] = df_london_lad_use_new['supergroup'].to_list()
    
    X.columns = X.columns.astype(str)
    y = df_london_lad_use_new[n].to_list()
    
# define the rf classifer model
    
    clf = RandomForestClassifier(max_depth=25, random_state=1)
    
    clf.fit(X, y)
    
# define the model evaluation procedure
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=15, random_state=1)
    
# evaluate the model and collect the scores
    n_scores = cross_val_score(clf, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
    
# report the model performance
    print('Mean Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
    fi_model = pd.DataFrame({'feature': df_london_lad_use_new.columns[1:11].to_list() + ['supergroup'],
                        'importance': clf.feature_importances_}).\
                        sort_values('importance', ascending = False)
    all_accuracy_list.append(n_scores)
    print(fi_model)
    # fi_model.to_csv(f'test/{n}.csv')
```
